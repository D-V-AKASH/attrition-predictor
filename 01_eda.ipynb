```python
attrition_count = df['Attrition'].value_counts()
print(attrition_count)
print(f" Attritioin rate : {round(attrition_count['Yes']/ len(df) *100,1)%"}

plt.figure(figsize=(6,4))
sns.countplot(data =df, x ='Attrition', palatte ='Set1')
plt.title("Attrition Distribution")
plt.ylabel("number of emps")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='OverTime', hue='Attrition', palette='Set2')
plt.title('Attrition by Overtime')
plt.ylabel('Number of Employees')
plt.show()
```

```python
print(pd.crosstab(df['OverTime'], df['Attrition'], normalize='index') * 100)
```

```python
plt.figure(figsize=(8,4))
sns.histplot(data=df, x='Age', hue='Attrition', bins=30, palette='Set2')
plt.title('Age Distribution by Attrition')
plt.ylabel('Number of Employees')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(data=df, x='MonthlyIncome', hue='Attrition', bins=30, palette='Set2')
plt.title('Monthly Income by Attrition')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='JobRole', hue='Attrition', palette='Set2')
plt.xticks(rotation=45, ha='right')
plt.title('Attrition by Job Role')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(data=df, x='YearsAtCompany', hue='Attrition', bins=20, palette='Set2')
plt.title('Years at Company by Attrition')
plt.show()
```

```python
plt.figure(figsize=(14,10))
correlation = df.select_dtypes(include=['int64','float64']).corr()
attrition_corr = correlation['Attrition' if 'Attrition' in correlation else correlation.columns[0]]

# Convert Attrition to number first
df['Attrition_num'] = df['Attrition'].map({'Yes': 1, 'No': 0})
correlation2 = df.select_dtypes(include=['int64','float64']).corr()

attrition_corr = correlation2['Attrition_num'].drop('Attrition_num').sort_values()

plt.figure(figsize=(8,10))
sns.barplot(x=attrition_corr.values, y=attrition_corr.index, palette='RdYlGn')
plt.title('Correlation of Every Feature with Attrition')
plt.xlabel('Correlation (-1 to +1)')
plt.axvline(x=0, color='black', linewidth=0.8)
plt.tight_layout()
plt.show()
```
